```python
import pandas as pd
from scipy import stats

# データの読み込み
df = pd.read_csv('MD2last.csv')

# 分析対象の列を抽出
x = df['日最大風速10m/s以上の日数']
y = df['日最大風速15m/s以上の日数']

# 線形回帰分析を実行して相関係数とp値を計算
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

print(f"相関係数 (r): {r_value:.4f}")
print(f"p値: {p_value:.4f}")

# p値の解釈
if p_value < 0.05:
    print("\np値は0.05より小さいため、この相関は統計的に有意です。")
else:
    print("\np値は0.05より大きいため、この相関は統計的に有意とは言えません。")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 日本語フォントの設定（Jupyter Notebook環境による）
plt.rcParams['font.family'] = 'Hiragino Sans'
plt.rcParams['axes.unicode_minus'] = False

# データの読み込み
df = pd.read_csv('MD2last.csv')

# 風速関連のカラムを選択
wind_columns = [
    '平均風速',
    '日最大風速10m/s以上の日数',
    '日最大風速15m/s以上の日数',
    '日最大風速20m/s以上の日数'
]

# 相関行列の計算
correlation_matrix = df[wind_columns].corr()

# ヒートマップの描画
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix,
            annot=True,        # 数値を表示
            cmap='coolwarm',     # カラーマップ
            center=0,          # 0を中心に
            vmin=-1, vmax=1,   # 範囲を-1から1に
            fmt=".2f",         # 小数点以下2桁で表示
            linewidths=.5)     # セルの境界線
plt.title('風速関連データ間の相関ヒートマップ')
plt.show()
```
