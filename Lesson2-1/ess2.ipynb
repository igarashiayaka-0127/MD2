```python
import pandas as pd

# CSVを読み込む
df = pd.read_csv('板垣.csv')

# timestampを0から始まる秒単位に変換
df['relative_time'] = (df['relative_time'] - df['relative_time'].iloc[0]) / 1000

# 保存
df.to_csv('板垣s.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSVファイルを読み込む
全 = pd.read_csv('全s.csv')
久我 = pd.read_csv('久我s.csv')
五十嵐 = pd.read_csv('五十嵐s.csv')
中里 = pd.read_csv('中里s.csv')
板垣 = pd.read_csv('板垣s.csv')
顧 = pd.read_csv('顧s.csv')

# X軸データをまとめる
data_to_plot = [
    全['AccX'].values,
    久我['AccX'].values,
    五十嵐['AccX'].values,
    中里['AccX'].values,
    板垣['AccX'].values,
    顧['AccX'].values
]

# Boxplot作成
plt.figure(figsize=(6, 4))
box = plt.boxplot(data_to_plot, labels=['全','久我','五十嵐','中里','板垣','顧'])

plt.xlabel('Person')
plt.ylabel('X-axis Acceleration (m/s²)')
plt.title('X-axis Data Distribution Comparison')
plt.grid(axis='y', alpha=0.3)

# 標準偏差を追加表示
std_values = [全['AccX'].std(), 久我['AccX'].std(), 五十嵐['AccX'].std(), 中里['AccX'].std(), 板垣['AccX'].std(), 顧['AccX'].std()]
for i, std in enumerate(std_values, 1):
    plt.text(i+0.3, plt.ylim()[1] * 0.65, f'σ={std:.3f}', ha='center')
# 最大値と最小値を追加表示
max_values = [全['AccX'].max(), 久我['AccX'].max(), 五十嵐['AccX'].max(), 中里['AccX'].max(), 板垣['AccX'].max(), 顧['AccX'].max()]
min_values = [全['AccX'].min(), 久我['AccX'].min(), 五十嵐['AccX'].min(), 中里['AccX'].min(), 板垣['AccX'].min(), 顧['AccX'].min()]
for i, (max_val, min_val) in enumerate(zip(max_values, min_values), 1):
    plt.text(i+0.3, plt.ylim()[1] * 0.85, f'Max={max_val:.3f}', ha='center', color='red')
    plt.text(i+0.3, plt.ylim()[0] * 0.95, f'Min={min_val:.3f}', ha='center', color='blue')    

plt.show()
```
